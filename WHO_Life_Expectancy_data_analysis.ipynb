```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.25, random_state = 0)
```

```python
#Training without feature scaling:
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

**Linear Regression**

```python

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print('Train Score: {:.4f}'.format(lr.score(X_train, y_train)))
print('Test Score: {:.4f}'.format(lr.score(X_test, y_test)))
```

```python
from sklearn import metrics
```

```python
  print("Mean Squared error : {:.4f}".format(metrics.mean_squared_error(y_test, y_pred)))
  print("R2 Score : {:.4f}".format(metrics.r2_score(y_test, y_pred)))
```

Scaling the features to a common range using MinMaxScaler and StandardScaler

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
std_scaler = StandardScaler()
mm_scaler = MinMaxScaler()
X_train_std = std_scaler.fit_transform(X_train)
X_test_std = std_scaler.fit_transform(X_test)
X_train_mm = mm_scaler.fit_transform(X_train)
X_test_mm = mm_scaler.fit_transform(X_test)
```

```python
lr_std = LinearRegression()
lr_std.fit(X_train_std, y_train)

lr_mm = LinearRegression()
lr_mm.fit(X_train_mm, y_train)

print("Standard scaler scaling train score : {:.6f}".format(lr_std.score(X_train_std, y_train)))
print("MinMax scaling train score : {:.6f}".format(lr_mm.score(X_train_mm, y_train)))
```

Both Scaling provide the same train scores of 0.7913 (equal to that without scaling)

```python
print("Standard scaler scaling test score : {:.4f}".format(lr_std.score(X_test_std, y_test)))
print("MinMax scaler scaling test score : {:.4f}".format(lr_mm.score(X_test_mm, y_test)))
```

**Decision Tree Regression**

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor()
dt.fit(X_train, y_train)

print("Train and test scores for decision tree regressor without scaling")
print('Train Score: {:.4f}'.format(dt.score(X_train, y_train)))
print('Test Score: {:.4f}'.format(dt.score(X_test, y_test)))
```

Adjusting max_depth hyperparameter to 6

```python
dt_2 = DecisionTreeRegressor(max_depth = 6)
dt_2.fit(X_train, y_train)

print('Train Score: {:.4f}'.format(dt_2.score(X_train, y_train)))
print('Test Score: {:.4f}'.format(dt_2.score(X_test, y_test)))
```

Decision tree regressor on scaled features

```python
dt_2.fit(X_train_std, y_train)
print('Standard scaler Train Score: {:.4f}'.format(dt_2.score(X_train_std, y_train)))
print('Standard scaler Test Score: {:.4f}'.format(dt_2.score(X_test_std, y_test)))
```

```python
dt_2.fit(X_train_mm, y_train)
print('MinMax scaler Train Score: {:.4f}'.format(dt_2.score(X_train_mm, y_train)))
print('MinMax scaler Test Score: {:.4f}'.format(dt_2.score(X_test_mm, y_test)))
```

Scaled feature show reduction in test scores:

Standard Scaler test score : 0.8856

MinMax Scaler test score : 0.8865

No Scaling test score for DecisionTreeRegressor : 0.9043

```python
#feature importance
pd.Series(dt_2.feature_importances_, index = X.columns).sort_values(ascending = False)
```

From above feature importance series, it can be inferred that countries having high prevalence of HIV/AIDS have lower life expectancy rates. Adult Mortality, Income composition of Resorces (can be equated to Schooling) and BMI (shows the nutrition status in population) also have large contribution in determining Life Expectancy rates

```python
#Moving ahead with prediction based on non-scaled Decision Tree Model:
dt_2.fit(X_train, y_train)
y_pred_dt = dt_2.predict(X_test)
print("Mean Squared Error for Decision Tree Model : {:.4f}".format(metrics.mean_squared_error(y_test, y_pred_dt)))
print("R2 score for Decision Tree Model : {:.4f}".format(metrics.r2_score(y_test, y_pred_dt)))
```

```python
rmse_dt = np.sqrt(metrics.mean_squared_error(y_test, y_pred_dt))
print("Root mean squared error for DT model : ", rmse_dt)
```

We get improved R2 score as well as mean squared error for decision tree regressor:

Decision tree Regression:

R2 score = 0.8879

MSE = 10.5039

Linear Regression:

R2 score = .7879

MSE = 19.8763



**Support Vector Machine Regression**

```python
from sklearn.svm import SVR
```

```python
svr = SVR()
svr.fit(X_train, y_train)

print('Train Score: {:.4f}'.format(svr.score(X_train, y_train)))
print('Test Score: {:.4f}'.format(dt_2.score(X_test, y_test)))
```

```python
svr_std = SVR()
svr_std.fit(X_train_std, y_train)

print('Standard Scaler Train Score: {:.4f}'.format(svr_std.score(X_train_std, y_train)))
print('Standard Scaler Test Score: {:.4f}'.format(svr_std.score(X_test_std, y_test)))
```

```python
svr_mm = SVR()
svr_mm.fit(X_train_mm, y_train)

print('MinMax Scaler Train Score: {:.4f}'.format(svr_mm.score(X_train_mm, y_train)))
print('MinMax Scaler Test Score: {:.4f}'.format(svr_mm.score(X_test_mm, y_test)))
```

We observe that scaling improves the training scores for Support Vector Regressor by a large amount. But not not improvement is seen in the test scores for both scaling. Standard Scaler regressor gives us the highest train score of 0.8672. We will predict based on the standard scaler scaled features.

```python
y_pred_svr = svr_std.predict(X_test_std)
print("Mean Squared Error : {:.4f}".format(metrics.mean_squared_error(y_test, y_pred_svr)))
print("R2 score : {:.4f}".format(metrics.r2_score(y_test, y_pred_svr)))
```

```python
rmse_svr = np.sqrt(metrics.mean_squared_error(y_test, y_pred_svr))
print("Root Mean Squared Error : {:.4f}".format(rmse_svr))
```
